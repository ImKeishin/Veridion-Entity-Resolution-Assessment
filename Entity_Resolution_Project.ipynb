```python
relevant_columns = ['company_name', 'main_country_code', 'main_region', 'main_city', 'main_postcode', 'main_latitude', 'main_longitude', 'locations', 
                    'company_type', 'year_founded', 'short_description', 'business_tags', 'business_model', 'product_type', 'main_sector', 'emails', 
                   'website_domain', 'sic_labels', 'status']
```

```python
# this isn't good displayment of the data that we want
relevant_data = []
for col in relevant_columns:
    relevant_data.append(df[col])
data = pd.DataFrame({'Column' : relevant_columns, 'Value' : relevant_data})
data.head(n=7)
```

```python
# this should be a better way to show it
data = df.copy()
for col in df.columns:
    if col not in relevant_columns:
        data.drop(col, axis=1, inplace=True)
data.head(n=6)
```

```python
data.info()
data['company_name'].fillna(value='-', inplace=True)
data['main_country_code'].fillna(value='-', inplace=True)
data['main_region'].fillna(value='-', inplace=True)
data['main_city'].fillna(value='-', inplace=True)
data['main_postcode'].fillna(value=np.nan, inplace=True)
data['main_latitude'].fillna(value=np.nan, inplace=True)
data['main_longitude'].fillna(value=np.nan, inplace=True)
data['locations'].fillna(value='-', inplace=True)
data['year_founded'].fillna(value=np.nan, inplace=True)
data.head(n=6)
```

```python
data['main_postcode'] = data['main_postcode'].astype(str).str.replace(r'\D', '', regex=True)
data['main_postcode'].replace('', '0', inplace=True)
data['main_postcode'] = data['main_postcode'].astype(int)
data = data.astype({'main_latitude' : float, 'main_longitude': float})
print(data['company_type'].value_counts())
print(data['status'].value_counts())
```

```python
data = data.drop('status', axis=1, inplace=True) # since status column has only 'Active' values meaning it's constant
```

```python
sns.countplot(y='company_type', data=data)
```

```python
# trying standardizing data but most of it is made of strings.
# I wanted to standardize only the main_postcode, main_latitude, main_longitude and maybe the main_country_code too.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_std = scaler.fit(data)
```

```python
data.head()
from sklearn.cluster import KMeans
inertia=[]
for k in range(1, 18000):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 18000), inertia, marker='o', markerfacecolor='r', markersize=8, lw=4)
plt.xlabel('Numar de clustere')
plt.ylabel('Inertia')
```
